## Import

```python
import torch
import numpy as np
import sys
import os
```

## Dataset & transoform

```python
sys.path.append(os.path.abspath(os.path.join("..")))
from utils.dataset import ImageCaptionDataset

from torchvision.transforms import RandomHorizontalFlip, Resize, RandomResizedCrop
from torchvision.transforms import Compose, ToTensor, Normalize


img_size = 224
transform = Compose(
    [
        Resize(size=(img_size, img_size)),
        RandomHorizontalFlip(p=0.5),
        ToTensor(),
        Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ]
)

dataset_train = ImageCaptionDataset(
    "../data/train", "../data/train/captions_train.txt", transform
)
```

## Visualize transoformations

```python
from utils.dataset import compare_transforms

dataset_no_transform = ImageCaptionDataset(
    "../data/train", "../data/train/captions_train.txt", Compose([ToTensor()])
)

data_transf_2 = ImageCaptionDataset(
    "../data/train",
    "../data/train/captions_train.txt",
    Compose(
        [
            RandomResizedCrop(size=(img_size, img_size)),
            RandomHorizontalFlip(p=0.5),
            ToTensor(),
            Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
        ]
    ),
)

compare_transforms([dataset_no_transform, data_transf_2, dataset_train], 5)
```
